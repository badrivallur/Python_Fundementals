```python
#dictionaries in python
```

```python
alien={'color':'green','start_position':'level1','points':5}
print(alien)
```

```python
print(alien['color'])
```

```python
for i,j in alien.items():
    print(i)
    print(j)
```

```python
for i,j in alien.items():
    print(f"key: {i}")
    print(f"value: {j}")
```

```python
for i,j in alien.items():
    print(f"key: {i},value: {j}")
```
